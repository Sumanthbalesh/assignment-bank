#  The Crank Nicolson Method

Yes! This is one of the well known finite difference method for numerical integration to the heat equations and also closely related partial differential equations.

$$ \frac {∂u}{∂t}= D \frac {∂^2u}{∂x^2}+f(u),$$  $$ \frac {∂u}{∂x}\mid_{x=0,L} =0 $$

Here, **u**-Concentration Variable, **x**- Space Variable, **D**-Diffusion co efficient of u, **f**-Reaction Term,**L**-Length of the one dimensional space domain.In this case Lets use Neumann Boundary conditions and specify that the solution u has zero space slope at the boundaries.

## Finite Difference Methods

The above reaction-diffusion equation describes the time evolution of variable **$u(x,t)$** in one space dimension (u is a line concentration). If we knew an analytic expression for **$u(x,t)$** then we could plot u in a two-dimensional coordinate system with axes t and **$x$**.To approximate **$u(x,t)$** numerically we discretize this two-dimensional coordinate system resulting, in the simplest case, in a two-dimensional regular grid.

After knowing a bit about finite difference methods let us discretize both time and space.

$$t_n = nΔt, n=0....., N-1,
x_j= jΔx, j=0,....,J-1$$

Here **$N$** is Number of discrete Time and **$J$** is Number of space points in grid. **$Δt$** is the time step and **$Δx$** is the space step.

$$ Δt=T/N,
Δx=L/J$$

where **$T$** is the point in time up to which we will integrate **$u$** numerically.

All we need to do is construct a numerical method that helps us to find an approximate solution of unknown analytic solution **$u(x,t)$**, that fits in reasonably well in these grid points.

We want construct a method that computes values **$U(jΔx,nΔt)$** *(note: capital **$U$**)* so that

$$ U(jΔx,nΔt) \approx\ u(jΔx,nΔt)$$

As a shorthand we will write **$U_j^n=U(jΔx,nΔt)$** and **$(j,n)$** to refer to grid point **$(jΔx,nΔt)$**.

# Now, Its time to draw Crank-Nicolson Stencil

First we shall construct a two dimensional grid and then we shall approximate the operators of our reaction diffusion system.

For example,if we want to approximate the time derivative which is on the left side in grid point **$(j,n)$**, we use the values of **$U$** in two specific grid points:

$$ \frac {∂u}{∂t}\mid_{x=jΔx, t=nΔt} \approx \frac {U_j^{n+1}-U_j^n}{Δt}$$

We can think of this scheme as a stencil that we superimpose on our **$(x,t)$**-grid and this particular stencil is commonly referred to as forward difference.

What form will the spatial term of Crank-Nicolson stencil take for spatial part of heat equation:

$$ \frac {∂^2u}{∂x^2}\mid_{x=jΔx,t=nΔt} \approx \frac {1}{2Δx^2}(U_{j+1}^n-2U_j^n+U_{j-1}^n+U_{j+1}^{n+1}-2U_j^{n+1}+U_{j-1}^{n+1})$$

To approximate $f(u(jΔx,nΔt))$ we write simply $f(U_j^n$).

<img src="./Image7.JPG">

Applying this stencil to grid point $(j,n)$ gives us the following approximation of our reaction-diffusion equation:

$$\frac {{U_j^{n+1}-U_j^n}}{Δt} =\frac {D} {2Δx^2} (U_{j+1}^n-2U_j^n+U_{j-1}^n+U_{j+1}^{n+1}-2U_j^{n+1}+U_{j-1}^{n+1}) + f(U_j^n).,$$

# How to reorderstencil into linear system:

In order to reorder our reaction diffusion equation, first let us define  σ.

$$ σ= \frac {DΔt}{2Δx^2}$$

After reordering

$$ - σU_{j-1}^{n+1}+(1+2 σ)U_j^{n+1}- σU{j+1}^{n+1}=  σU_{j-1}^n+(1-2 σ)U_j^n+ σU_{j+1}^n+Δtf(U_j^n)$$

WOW! Isnt this amazing. But there is something wrong.This equation makes sense for space indices $j=1,…,J−2$ but it does not make sense for indices $j=0$ and $j=J−1$ (on the boundaries):

$$ j=0:  -σU_{-1}^{n+1}+(1+2σ)U_0^{n+1}-σU_1^{n+1}=σU_{-1}^n+(1-2σ)U_0^n+σU_1^n+Δtf(U_0^n),$$  $$ j=J-1: -σU_{J-2}^{n+1}+(1+2σ)U_{J-1}^{n+1}-σU_J^{n+1}=σU_{J-2}^n+(1-2σ)U_{J-1}^n+σU_J^n+Δtf(U_{J-1}^n)$$

The problem here is that the values $U_{-1}^n$ and $U_J^n$ lie outside our grid.

If we consider the Neumann boundary condition, we can work out what these values should be equal to. Now is the time to discretize! We discretize the boundary condition at $j=0$ with backward difference and at $j=J-1$ with forward difference.

$$\frac {{U_1^n-U_0^n}}{Δx} =0,$$   $$\frac {{U_J^n-U_{J-1}^n}}{Δx}=0 $$

These two equations make it clear that we need to amend our above numerical approximation for **$j=0$** with the identities $U_0^n=U_1^n$ and $U_0^{n+1}=U_1^{n+1}$, and for $j=J−1$ with the identities $U_{J−1}^n=U_J^n$ and $U_{J−1}^{n+1}=U_J^{n+1}$.

Let us reinterpret our numerical approximation of the line concentration of u in a fixed point in time as a vector $U_n$:

We found this notation. Is this helpful? Yes! it is. By using this we can now write our above approximation for a fixed point in time, $t=nΔt$, compactly as a linear system:

<img src="./Image15.JPG">

We have a well defined initial condition at $n=0$ from which our numerical integration starts,$U^0$, the vector $U^{n+1}$ on the left hand side is the only unknown in this system of linear equations

Thus, to integrate numerically our reaction-diffusion system from time point $n$ to $n+1$ we need to solve numerically for vector $U^{n+1}$

The matrix on the left hand side looks huge, lets give it a name. Hmmmm Let it be called $A$.On the right hand side lets name it $B$ and the vector on the right side $f^n$.

Considering this notation we can write another equation which will be of the form:

$$AU^{n+1} = BU^n + f^n$$

In this linear equation, matrices $A$ and $B$ are defined by our problem: we need to specify these matrices once for our problem and incorporate our boundary conditions in them. Vector $f^n$ is a function of $U^n$ and so needs to be reevaluated in every time point $n$. We also need to carry out one matrix-vector multiplication every time point, $BU^n$, and one vector-vector addition, $BU^n+f^n$.

Note a point, Inversion of matrix $A$ is most expensive numerical equation which is used to solve $U^{n+1}$, however we may get away with doing this only once and store the inverse of $A$ as $A^{−1}$.

$$U^{n+1}= A^{-1}(BU^n+f^n)$$

# We learnt many things, wow! thats amazing. Now its time to do an example.

Let us apply the CN method to a two-variable reaction-diffusion system.

$$\frac {∂u}{∂t}=D_u \frac {∂^2u}{∂x^2}+f(u,v),$$      $$\frac {∂v}{∂t}=D_v \frac{∂^2v}{∂x^2}−f(u,v),$$

With Neumann Boundary conditions:

$$\frac {∂u}{∂x}\mid_{x=0,L}=0,$$     $$\frac {∂v}{∂x}\mid_{x=0,L}=0.$$

$u$ and $v$ are the variables of this system which represents the concentration of active form and concentration of inactive form respectively. The reaction term $f(u,v)$ describes the activation and inactivation between these two states of protien.

### This system has a particular property, which is, "The inactive form has greater diffusivity than the active form" ie $D_v\gg D_u.$

We need to setup two approximations by using CN method to integrate this system numerically

$$ A_uU^{n+1}=B_uU^n+f^n,$$   $$ A_vV^{n+1}=B_vV^n-f^n.$$

This leaves us two σ terms, $σ_u= \frac {D_uΔt}{2Δx^2}$ and $σ_v=\frac {D_vΔt}{2Δx^2} $

#### We shall make use of Python library NumPy for Matrix vector multiplication, vector-vector addition and matrix inversion. Pyplot is used to visualize our numerical solutions.

```python
import numpy
from matplotlib import pyplot
```

Numpy allows us to truncate the numerical values of matrices and vectors to improve their display with set_printoptions.

```python
numpy.set_printoptions(precision=3)
```

# Lets now Specify Grid

The one dimensional domain that we are using has unit length and lets define $J=100$ equally spaced grid points in this domain. By doing this we divide our domain into $J-1$ subintervals, each of length $dx$.

```python
L = 1.
J = 100
dx = float(L)/float(J-1)
x_grid = numpy.array([j*dx for j in range(J)])
```

Now we define $N = 1000$ equally spaced grid points on our time domain of length $T = 200$ thus dividing our time domain into $N-1$ intervals of length $dt$.

```python
T = 200
N = 1000
dt = float(T)/float(N-1)
t_grid = numpy.array([n*dt for n in range(N)])
```

# Specify System Parameters and the Reaction Term

Lets choose the parameter values based on [ Bistable Reaction-Diffusion System](http://www.sciencedirect.com/science/article/pii/S0006349508704442)

```python
D_v = float(10.)/float(100.)
D_u = 0.01 * D_v

k0 = 0.067
f = lambda u, v: dt*(v*(k0 + float(u*u)/float(1. + u*u)) - u)
g = lambda u, v: -f(u,v)
 
sigma_u = float(D_u*dt)/float((2.*dx*dx))
sigma_v = float(D_v*dt)/float((2.*dx*dx))

total_protein = 2.26
```

# Now, its time to Specify the initial condition

As discussed in [ Bistable Reaction-Diffusion System](http://www.sciencedirect.com/science/article/pii/S0006349508704442), we can see some interesting behaviour in the steady state of this system  if we choose a heterogeneous initial condition for $u$.

Here, we initialize u with a step-like heterogeneity:

```python
no_high = 10
U =  numpy.array([0.1 for i in range(no_high,J)] + [2. for i in range(0,no_high)])
V = numpy.array([float(total_protein-dx*sum(U))/float(J*dx) for i in range(0,J)])
```

We make certain that total protein amounts equal a certain value, total_protein. The importance of this was discussed by [ Bistable Reaction-Diffusion System](http://www.sciencedirect.com/science/article/pii/S0006349508704442)

### Its time to plot our initial condition for confirmation:

```python
ylim((0., 2.1))
xlabel('x'); ylabel('concentration')
pyplot.plot(x_grid, U)
pyplot.plot(x_grid, V)
pyplot.show()
```

The blue curve is the initial condition for $U$, stored in Python variable $U$, and the green curve is the initial condition for $V$ stored in $V$.

# What next? Lets Create Matrices

The matrices that we need to construct are all tridiagonal so they are easy to construct with numpy.diagflat.

```python
A_u = numpy.diagflat([-sigma_u for i in range(J-1)], -1) +\
      numpy.diagflat([1.+sigma_u]+[1.+2.*sigma_u for i in range(J-2)]+[1.+sigma_u]) +\
      numpy.diagflat([-sigma_u for i in range(J-1)], 1)
        
B_u = numpy.diagflat([sigma_u for i in range(J-1)], -1) +\
      numpy.diagflat([1.-sigma_u]+[1.-2.*sigma_u for i in range(J-2)]+[1.-sigma_u]) +\
      numpy.diagflat([sigma_u for i in range(J-1)], 1)
        
A_v = numpy.diagflat([-sigma_v for i in range(J-1)], -1) +\
      numpy.diagflat([1.+sigma_v]+[1.+2.*sigma_v for i in range(J-2)]+[1.+sigma_v]) +\
      numpy.diagflat([-sigma_v for i in range(J-1)], 1)
        
B_v = numpy.diagflat([sigma_v for i in range(J-1)], -1) +\
      numpy.diagflat([1.-sigma_v]+[1.-2.*sigma_v for i in range(J-2)]+[1.-sigma_v]) +\
      numpy.diagflat([sigma_v for i in range(J-1)], 1)
```

Do we have any idea as how our A_u looks like. Lets give it a try

```python
print A_u
```

# How shall we solve this system iteratively?

Now inorder to advance our system by one time step we need to do one matrix vector multiplication followed by one vector vector addition on the right hand side

Now, let us rewrite our reaction term so that it accepts concentration vectors $U^n$ and $V^n$ as arguements and return vector $f^n$.

Our non vectorial definition of $f$ is as follows

$$f = \lambda u, v: v*(k_0 + float(u*u)/float(1. + u*u)) - u$$

```python
f_vec = lambda U, V: numpy.multiply(dt, numpy.subtract(numpy.multiply(V, 
                     numpy.add(k0, numpy.divide(numpy.multiply(U,U), numpy.add(1., numpy.multiply(U,U))))), U))
```

Now, lets make sure that this produces same values as our non-vectorial $f$

```python
print f(U[0], V[0])
```

```python
print f_vec(U, V)
```

Accounting for rounding of the displayed values due to the set_printoptions we set above, we can see that $f$ and $f_{vec}$ generate the same values for our initial condition at both ends of our domain.

We will use numpy.linalg.solve to solve our linear system each time step.

While we integrate our system over time we will record both $U$ and $V$ at each time step in $U_{record}$ and $V_{record}$ respectively so that we can plot our numerical solutions over time.

```python
U_record = []
V_record = []

U_record.append(U)
V_record.append(V)

for ti in range(1,N):
    U_new = numpy.linalg.solve(A_u, B_u.dot(U) + f_vec(U,V))
    V_new = numpy.linalg.solve(A_v, B_v.dot(V) - f_vec(U,V))
    
    U = U_new
    V = V_new
    
    U_record.append(U)
    V_record.append(V)
```

# We are almost there. Now its time to plot our Numerical Solution

```python
ylim((0., 2.1))
xlabel('x'); ylabel('concentration')
pyplot.plot(x_grid, U)
pyplot.plot(x_grid, V)
pyplot.show()
```

And here is a kymograph of the values of U. This plot shows concisely the behaviour of U over time and we can clear observe the wave-pinning behaviour described by [ Bistable Reaction-Diffusion System](http://www.sciencedirect.com/science/article/pii/S0006349508704442). Furthermore, we observe that this wave pattern is stable for about 50 units of time and we therefore conclude that this wave pattern is a stable steady state of our system.
